- Class TiendaOnline: es la clase o plantilla.
- El objeto: es lo que voy a crear, es la tienda.
- Atributos: son las características q le damos al objeto (aqui son nombre, precio, cantidad).
- Método: son las funciones de la clase y generalmente son acciones y se utilizan verbos.

```python
class TiendaOnline:                                                         # esta es la plantilla
    def __init__(self):
        self.inventario = []                                                #empezamos con la lista vacia
        self.ventas_totales = 0
        self.clientes = {}                                                  #Bonus, añadimos atribito dicc clientes                 

    def agregar_producto(self, nombre, precio, cantidad):                   # estos son los metodos
                                         
        for producto in self.inventario:                                    #1. agregar prod. para iterar utilizamos el for
            if producto["nombre"] == nombre:                                #compruebo si el producto esta y si hay q agragarle cantidad 
                producto["cantidad"] += cantidad                            #accediendo primero a la cantidad y añadiendo el producto
                print("Producto encontrado y cantidad añadida.")
                break                                                       #si pongo return no me ejecutaria el else, no se necesitaria, con el break si me salta al else

        else:    
            nuevo_producto = {"nombre": nombre, "precio": precio, "cantidad": cantidad}    # añadimos el nuevo producto en forma dicc
            self.inventario.append(nuevo_producto)                                         # utilizamos append porque es una lista, en diccionario sería update 
            print("¡El nuevo producto se ha añadido correctamente!")
            
    
    def ver_inventario(self):                                                 #2. ver inventario (es una lista).  
        print("inventario:")                                                  #el 2º print en formato comillas doble fuera y simple dentro, para evitar errores
        for producto in self.inventario:                                      # accediendo a claves de diccionario
            print(f"Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")

    
    def buscar_producto(self, nombre):                                        #3. buscar producto en inventario por nombre
        encontrado = False

        for producto in self.inventario:                                       #hay que buscar en el inventario, y para no hacerlo como antes utilizo true y false,
            if producto["nombre"] == nombre:                                   
                print(f"Producto encontrado, Nombre: {producto['nombre']}, Precio: {producto['precio']}, Cantidad: {producto['cantidad']}")
                encontrado = True

        if not encontrado:                                                      # si el prod no existe se envía mensaje informando
            print("Producto no encontrado.")

    
    def actualizar_stock(self, nombre, cantidad):                               #4. actualizar stock de producto en inventario. 
        for producto in self.inventario:                
            if producto['nombre'] == nombre:                                    # primero busco si existe y si esta sumo al stock lo que me den. parto de cero
                if producto ['cantidad'] + cantidad < 0:                        #podria darme un negativo si es una devolucion, y le aviso que no hay stock suficiente
                   print("¡No hay suficiente stock para la cantidad que solicita!")      # en el caso que corresponda se añade o resta la cantidad que me den, al stock
                else:
                    producto['cantidad'] += cantidad
                    print(f"¡Stock actualizado correctamente!. Nombre: {producto['nombre']}, Cantidad: {producto['cantidad']}")
                break
        else:
            print("Producto no encontrado.")

    
    def eliminar_producto(self, nombre):                                        #5.eliminar producto del inventario por nombre.
        for producto in self.inventario:                                        #recorremos con for para buscar
            if producto['nombre'] == nombre:                                    #eliminamos con remove xq quita el prod concreto que me digan (zapatos) del inventario
                self.inventario.remove(producto)                                # y que corresponde a la clave nombre
                print("Producto eliminado correctamente.")
                break
        else:
            print("No tenemos ese producto en nuestro stock.")


    def calcular_valor_inventario(self):                                         #6.calcular y mostrar el valor total del inventario
        valor_total = 0                                                          #creo un acumulador para meter elvalor total
        for producto in self.inventario:                
            valor_total += producto['precio']  * producto['cantidad']            # para averiguarlo * el precio por la cantidad que tenemos
        print(f"Este es el valor total del inventario: {valor_total} €")         # y la suma la meto en el valor total. (valor_total = valor_total + (precio * cantidad))

                                                                # **Bonus, 1.agregar clientes**
    def agregar_cliente(self, nombre, email):                                       
        if nombre in self.clientes:
            print(f"El Cliente: {nombre} con correo: {email} ya esta registrado.")              #primero revisamos si el cliente esta, para evitar duplicados
        else:
            self.clientes[nombre] = {"email": email, "compras": []}                              #aqui creo la nueva entrada vacia para cuando me compre el clte
            print(f"El Cliente: {nombre} con correo: {email}, Ha sido registrado con éxito.")
        
    
    def ver_clientes(self):                                                                       
        for nombre, datos in self.clientes.items():                                             # # 2. mostrar el listado de clientes registrados (es dicc),lo hago con items donde para ese cliente me da su clave y valor correspodeinte.     
            print(f"Estos son los clientes registrados: Nombre: {nombre}, datos: {datos}")      # debo acceder como dicc añadiendo datos para des empaquetarlo, para clienta Ingrid dame sus datos: q son email y compras
                                                                                                # si meto print(self.clientes) estaria sacando la info 2veces
    
    def realizar_compra(self):                                                                  # 3. Realizar compra
        carrito = {}

        while True:

            producto_cliente = input("Escribe 'ver' para ver productos o 'salir' para abandonar compra").lower().strip()
            if producto_cliente == "salir":
                print("¡hasta la próxima!")
                break
            elif producto_cliente == "ver":
                self.ver_inventario()
                producto_cliente = input("Escribe el nombre del producto que quieres comprar: ").lower().strip()
                continue
            
            if product_cliente not in self.inventario:
               print("Producto no disponibe. Elige uno de la lista")
            

            
            
```
