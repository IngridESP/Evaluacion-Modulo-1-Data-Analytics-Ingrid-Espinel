````python
class TiendaOnline:     # es la plantilla
    def __init__(self):
        self.inventario = []                         #empezamos con la lista vacia
        self.ventas_totales = 0                      

    def agregar_producto(self, nombre, precio, cantidad):       # estos son los metodos
                                         
        for producto in self.inventario:                     #1. agregar prod. para iterar utilizamos el for
            if producto["nombre"] == nombre:                   #compruebo si el producto esta y si hay q agragarle cantidad 
                producto["cantidad"] += cantidad               #accediendo primero a la cantidad y añadiendo el producto
                print("Producto encontrado y cantidad agregada")
                break                                           #si pongo return no me ejecutaria el else, no se necesitaria, con el brak si me salta alelse

        else:    
            nuevo_producto = {"nombre": nombre, "precio": precio, "cantidad": cantidad}      #añadimos el nuevo producto en forma dicc
            self.inventario.append(nuevo_producto)                                          #utilizamos append porq es una lista, en dicc es update 
            print("el nuevo producto se ha añadido correctamente")
            
    
    def ver_inventario(self):                                    #2. ver inventario   
        print("inventario:")                                     #el 2º print en formato comillas doble fuera y simple dentro
        for producto in self.inventario:                            #accediendoa claves de dicc
            print(f"nombre: {producto['nombre']}, Precio: ${producto['precio']}, cantidad: {producto['cantidad']}")

    
    def buscar_producto(self, nombre):                          #3. buscar prod en inventario por nombre
        encontrado = False

        for producto in self.inventario:        #hay que busca r en el inventario parano ahcerlo como antes utilizo true y false,
            if producto["nombre"] == nombre:        ## pero aqui nose rompe como con el break.
                print(f"Producto encontrado. nombre: {producto['nombre']}, Precio: {producto['precio']}, cantidad:{producto['cantidad']}")
                encontrado = True

        if not encontrado:                             # si el prod no existe seenviamensaje informando
            print("Producto no encontrado")

    
    def actualizar_stock(self, nombre, cantidad):       #4. actualizar stock de prod en inventario. 
        for producto in self.inventario:                
            if producto['nombre'] == nombre:                # primero busco si existe y si esta sumo al stock lo que den. parto de cero
                if producto ['cantidad'] + cantidad < 0:    #podria darme un engativo si es una devolucion, y le aviso que no hay stock suficiente
                   print("No hay suficiente stock")         # en caso añado la cantidad que me den al stock
                else:
                    producto['cantidad'] += cantidad
                    print(f"¡Stock actualizado correctamente!. Nombre: {producto['nombre']}, cantidad: {producto['cantidad']}")
                break
        else:
            print("Producto no encontrado")

    
    def eliminar_producto(self, nombre):                #5.eliminar prod del inventario por nombre.
        for producto in self.inventario:                   #recorremos con for para buscar
            if producto['nombre'] == nombre:                #eliminamos con remove xq quita el prod que me digan (camiseta)concreto del inventario
                self.inventario.remove(producto)            # que corresponde a la clave nombre
                print("Producto eliminado correctamente")
                break
        else:
            print("No tenemosese producto.")

    def calcular_valor_inventario(self):                #6.calcular y mostrar valor totaldel inventario
        valor_total = 0                                 #creo un acumulador para meter elvalor total
        for producto in self.inventario:                
            valor_total += producto['precio']  * producto['cantidad']         # para averiguarlo * el precio por la cantidad que tenemos
        print(f"Este es el valor total del inventario: {valor_total} €")   # y la suma la meto en el valor total. (valor_total = valor_total + (precio * cantidad))





````
